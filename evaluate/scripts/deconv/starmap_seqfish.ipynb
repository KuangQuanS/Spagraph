```python
import scanpy as sc
import pandas as pd

counts = pd.read_csv('/mnt/d/ST_Graduation_Project_data/database/STARmap/combined_spatial_count.txt',sep='\t')
print(counts)
location = pd.read_csv('/mnt/d/ST_Graduation_Project_data/database/STARmap/combined_Locations.txt',sep='\t')
print(location)

adata = sc.AnnData(counts.values)
adata.var_names = counts.columns
print(adata.var_names)
adata.obsm['spatial'] = location.values
adata.write_h5ad('/mnt/d/ST_Graduation_Project_data/database/STARmap/STARmap_SP.h5ad')
```

```python
import scanpy as sc
import pandas as pd


counts = pd.read_csv('/mnt/d/ST_Graduation_Project_data/database/STARmap/Spatial_count.txt',sep='\t')
print(counts)
location = pd.read_csv('/mnt/d/ST_Graduation_Project_data/database/STARmap/Locations.txt',sep='\t')
print(location)

adata = sc.AnnData(counts.values)
print(adata.obs_names)
adata.obsm['spatial'] = location.values

meta = pd.read_csv('/mnt/d/ST_Graduation_Project_data/database/STARmap/Spatial_annotate.txt', sep='\t',index_col=0)
print(meta)

adata.obs = meta
print(adata.obs)
print(adata.obs['celltype'].value_counts())
# 把 celltype 设成 category（画图更好看）
adata.obs['celltype'] = adata.obs['celltype'].astype('category')

sc.pl.embedding(
    adata,
    basis='spatial',
    color='celltype',
    s=50,           # 点大小
    alpha=0.8,      # 透明度
)
```
