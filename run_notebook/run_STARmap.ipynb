```python
import spagraph as spg

sc_file = f"/home/maweicheng/ST_data/seqFISH+/scRNA.h5ad"
st_file = f"/home/maweicheng/ST_data/seqFISH+/Spatial.h5ad"
output_dir = f"/home/maweicheng/ST_data/seqFISH+"

art = spg.vae(sc_file=sc_file, st_file=st_file, resolution=4, top_n_per_type=100, output_dir=output_dir, precomputed_marker_file=f"/home/maweicheng/ST_data/seqFISH+/final_genes.txt")

# art = spg.vae(sc_file=sc_file, st_file=st_file, top_n_per_type=100, marker_selection_method="l1", output_dir=output_dir)
res = spg.deconv(vae=art, output_dir=output_dir, k_cells_per_cluster=15, print_every=25, scale_basis="marker", lambda_mse=1, save_all_trials=True, save_reconstructed_genes=True) 
```
